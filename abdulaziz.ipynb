```python
#Q No 5 Write a program to identify duplicate values from list.

a = ['xyz','hamza','xyz','hamza',2,45,12,45]
b=[]
for i in a:
    if i not in b:
        b.append(i)
    else:
        print(str(i) + " is duplicated")
```
